```python
# Methode 1
# Load the AFINN-111 sentiment lexicon
def load_afinn():
    afinn = {}
    with open("AFINN-111.txt", "r") as afinn_file:
        for line in afinn_file:
            word, score = line.strip().split("\t")
            afinn[word] = int(score)
    return afinn
```

```python
# Detect sentiment of a single word using AFINN-111
def detect_sentiment(word, afinn_lexicon):
    sentiment_score = afinn_lexicon.get(word, 0)
    if sentiment_score > 0:
        return 1
    elif sentiment_score < 0:
        return -1
    else:
        return 0
```

```python
def liste_zeros(longueur):
    liste = [0] * longueur
    return liste
```

```python
i=0
# dataset2=dataset.head(10)
sentiment_scores1 = liste_zeros(len(dataset))
afinn_lexicon = load_afinn()
for sentence in dataset['Clean'] :
  for word in sentence:
    sentiment_scores1[i] = detect_sentiment(word, afinn_lexicon)+sentiment_scores1[i]
    # print(f"The sentiment of '{word}' is {sentiment_scores1[i]}")
  i=i+1
  # print(f"The sentiment of '{sentence}' is {sentiment_scores1}")
```

```python
# Encoding
from sklearn import preprocessing
custom_encoding = { 'Neutral':0, 'Negative': 1, 'Positive': 2}
label_encoder = preprocessing.LabelEncoder()
label_encoder.fit(dataset['Status'].map(custom_encoding))
dataset['Target'] = label_encoder.transform(dataset['Status'].map(custom_encoding))
```

```python
# Dictionary creation from the dataset (for model training/prediction)
data_dict = {
    'Feature': dataset['Text '].tolist(),
    'Target': dataset['Target'].tolist()
}
```

```python
# Data split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data_dict['Feature'][:100000], data_dict['Target'][:100000], test_size=0.2)
```

```python
# Feature extraction: Bag of Words representation
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
x_train_vec = vectorizer.fit_transform(x_train)
x_test_vec = vectorizer.transform(x_test)
```
